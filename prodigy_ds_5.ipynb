```python
counts = df['Weather_conditions'].value_counts()
plt.pie(counts.values, labels=counts.index, startangle=140)
plt.title('Distribution of Weather Condition')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Sex_of_driver', data=df, palette= 'pastel', hue='Sex_of_driver', legend=False)
plt.xlabel('Sex of Driver')
plt.ylabel('Count')
plt.title('Distribution of Accidents by Driver Sex')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Accident_severity', data=df, palette= 'pastel', hue='Accident_severity', legend=False)
plt.xlabel('Accident severity')
plt.ylabel('Count')
plt.title('Distribution of Accident severity')
```

```python
education_counts = df['Educational_level'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(education_counts.values, labels=education_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Accident by Education level')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='Educational_level', data=df, hue='Educational_level', palette='pastel', legend=False)
plt.title('Distribution by Education level')
plt.xlabel('Count')
plt.ylabel('Educational level')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='Road_surface_conditions', data=df, hue='Road_surface_conditions', palette='pastel', legend=False)
plt.title('Distribution by Road conditions')
plt.xlabel('Count')
plt.ylabel('Road surface conditions')
plt.show()
```
